
 
## Correction - Exercice 5 : Fonctions convexes



## 🔹 Rappels mathématiques

### 1. Fonction convexe
Une fonction $f$ est convexe si :
$$ f(\lambda x_1 + (1-\lambda)x_2) \leq \lambda f(x_1) + (1-\lambda)f(x_2) $$

### 2. Interprétation géométrique
Le segment entre deux points du graphe est **au-dessus** de la courbe.

### 3. Fonctions données
- $f(x) = x^2$
- $g(x,y) = \log(e^x + e^y)$


## 🔹 1) Tracer la courbe f(x) = x²

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
f = x**2

plt.figure()
plt.plot(x, f)
plt.title("Courbe de f(x) = x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

## 🔹 2) Points et segment (convexité)

```python

# Points
x1, x2 = 2, -3
y1, y2 = x1**2, x2**2

plt.figure()
plt.plot(x, f)

# Points rouges
plt.scatter([x1, x2], [y1, y2], color='red')

# Segment vert
plt.plot([x1, x2], [y1, y2], color='green')

plt.title("Illustration de la convexité")
plt.grid()
plt.show()
```

## 🔹 3) Fonction g(x,y)

```python

from mpl_toolkits.mplot3d import Axes3D

def g(x, y):
    return np.log(np.exp(x) + np.exp(y))

# Grille
X = np.linspace(-4, 4, 50)
Y = np.linspace(-4, 4, 50)
X, Y = np.meshgrid(X, Y)
Z = g(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

# Points
p1 = (2, 3, g(2,3))
p2 = (1, -3, g(1,-3))

# Points rouges
ax.scatter(*p1, color='red')
ax.scatter(*p2, color='red')

# Segment cyan
ax.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]], color='cyan')

ax.set_title("Surface de g et segment")
plt.show()
```
