 
## Correction - Exercice 6 : Étude des points critiques



## 🔹 Rappels mathématiques

### 1. Points critiques
Les points critiques sont les solutions de :
$$ f'(x) = 0 $$

### 2. Dérivée seconde
- Si $f''(x) > 0$ → minimum local  
- Si $f''(x) < 0$ → maximum local  
- Si $f''(x) = 0$ → point de selle (ou test non concluant)

### 3. Fonction étudiée
$$ f(x) = x^3 - 3x^2 + 2 $$


## 🔹 1) Définition de la variable et de la fonction

```python

import sympy as sp

# Définition de la variable réelle
x = sp.symbols('x', real=True)

# Définition de la fonction
f = x**3 - 3*x**2 + 2

f
```

## 🔹 2) Calcul de la dérivée première

```python

f_prime = sp.diff(f, x)

print("f'(x) =")
sp.pprint(f_prime)
```

## 🔹 3) Résolution de f'(x) = 0

```python

critical_points = sp.solve(f_prime, x)

print("Points critiques :")
print(critical_points)
```

## 🔹 4) Calcul de la dérivée seconde

```python

f_second = sp.diff(f_prime, x)

print("f''(x) =")
sp.pprint(f_second)
```

## 🔹 5) Nature des points critiques

```python

for point in critical_points:
    value = f_second.subs(x, point)

    print(f"Analyse au point x = {point}:")

    if value > 0:
        print("→ Minimum local")
    elif value < 0:
        print("→ Maximum local")
    else:
        print("→ Point de selle ou test non concluant")
```
