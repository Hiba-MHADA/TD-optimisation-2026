## 1. Définition des fonctions  

## Soit les fonctions réelles :

\[
\begin{aligned}
f(x) &= 2x^2 - 3, \\
g(x) &= x^3 - 3x^2 + 2, \\
h(x) &= x^5 + 4x^3 + 4x.
\end{aligned}
\]

```python
# TODO Import Sympy as sp
import sympy as sp
from IPython.display import display, Math

# ==========================
# 📌 Définir la variable
# ==========================
# TODO : définir x comme variable réelle
x = sp.symbols('x', real=True)

# ==========================
# 📌 Définir les fonctions
# ==========================

# TODO : f(x) = 2x^2 - 3
f = 2*x**2 - 3 

# TODO : g(x) = x^3 - 3x^2 + 2
g = x**3 - 3*x**2 + 2

# TODO : h(x) = x^5 + 4x^3 + 4x
h = x**5 + 4*x**3 + 4*x

# ==========================
# 📌 Affichage des résultats
# ==========================

print("📌 Les trois fonctions sont :\n")

display(Math(r"f(x) = " + sp.latex(f)))
# TODO : Afficher les autres fonctions
display(Math(r"g(x) = " + sp.latex(g)))
display(Math(r"h(x) = " + sp.latex(h)))
```

## 2. Calcul de la dérivée première de f

```python
# ==========================
# 📌 Calcul de la dérivée
# ==========================
# TODO : calculer la dérivée de f par rapport à x
df = sp.diff(f, x)

# ==========================
# 📌 Affichage du résultat
# ==========================

display(Math(r"f'(x) = " + sp.latex(df)))
```

## 7. Tracé de h'(x)

```python
# ==========================
# 📌 Variable symbolique et fonction
# ==========================
# TODO : définir x comme variable réelle SymPy
x = sp.symbols('x', real=True)

# TODO : définir la fonction h(x)
h = x**5 + 4*x**3 + 4*x

# ==========================
# 📌 Dérivée
# ==========================
# TODO : calculer la dérivée dh = h'(x)
dh = sp.diff(h, x)

# ==========================
# 📌 Domaine numérique
# ==========================
# TODO : générer X entre -3 et 3 (400 points)
X = np.linspace(-3, 3, 400)

# ==========================
# 📌 Évaluation numérique de la dérivée
# ==========================
# TODO : calculer Ydh = dh(x) pour chaque valeur de X
dh_func= sp.lambdify(x, dh, 'numpy')
Ydh = dh_func(X)


# ==========================
# 📌 Tracé
# ==========================
# TODO : tracer la dérivée
plt.plot(X, Ydh)

# ==========================
# 📌 Mise en forme
# ==========================
plt.xlabel('x')
plt.ylabel("h'(x)")
plt.title("Dérivée de h(x)")
plt.grid(True)

# TODO : afficher le graphique
plt.show()
```
